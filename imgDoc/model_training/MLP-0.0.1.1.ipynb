```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import pickle 
```

```python

from sklearn.datasets import load_iris
# Loading dataset
iris = load_iris()  
X, y = iris.data, iris.target
len(X), X, y
```

```python
keys = list(set(dataset["y"]))
Y_keys = dict()
for i, key in enumerate(keys):
    zeros = np.zeros((3))
    zeros[i] = 1
    Y_keys[key] = zeros
Y_keys
Y = np.array([Y_keys[yi] for yi in dataset["y"]])
Y
```

```python
print("0:", len([yi for yi in Y if yi == 0]))
print("1:", len([yi for yi in Y if yi == 1]))
print("2:", len([yi for yi in Y if yi == 2]))
```

```python
X = np.array(dataset["x"])
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42)
```

```python
# Creating (MLP) classifier
clf = MLPClassifier(hidden_layer_sizes=(8, 4, 8), max_iter=1000,
                    random_state=42)
```

```python
# Training the model
clf.fit(X_train, y_train)
# Making prediction
y_pred = clf.predict(X_test)  
```

```python
# Determining Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# save the model 
filename = 'model-1.sav'
pickle.dump(clf, open(filename, 'wb')) 
  
# load the model 
load_model = pickle.load(open(filename, 'rb')) 
```

```python
load_model.predict(np.array(X_train))  
```
