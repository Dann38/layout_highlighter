```python
import json
with open("./dataset/json/mini_publaynat_mlp_len_ang_veclen50.json") as f:
    dataset = json.load(f)

print("Пример, x:", dataset["x"][0])
print("Размер x:", len(dataset["x"][1]))
print("Пример, y:", dataset["y"][1])
print("Кол-во записей:", len(dataset["x"]))
```

```python
import numpy as np
import pandas as pd
from time import time

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import pickle 
```

```python
keys = list(set(dataset["y"]))
Y_keys = dict()
for i, key in enumerate(keys):
    zeros = np.zeros((len(keys)))
    zeros[i] = 1
    Y_keys[key] = zeros
Y_keys
Y = np.array([Y_keys[yi] for yi in dataset["y"]] )
X = np.array(dataset["x"])
```

```python
x1 = [x for i, x in enumerate(X) if Y[i][0] == 1]
y1 = [y[:3] for y in Y if y[0] == 1]
x2 = [x for i, x in enumerate(X) if Y[i][1] == 1]
y2 = [y[:3] for y in Y if y[1] == 1]
x3 = [x for i, x in enumerate(X) if Y[i][2] == 1]
y3 = [y[:3] for y in Y if y[2] == 1]
```

```python
len(x3)
```

```python
len_text = 955
len_header = 995
len_list = 995

X = x1[:len_text]+ x2[:len_header] + x3[:len_list]
Y = y1[:len_text]+ y2[:len_header] + y3[:len_list]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42)
```

```python
i = 0
df = pd.DataFrame(columns = ['alpha','max_iter','train_acc','test_acc','train_time'])
for a in [0.00001,0.0001,0.001,0.01, 0.1, 1, 10]:
    for mi in [10,100,200,500,1000,2000]:
        st = time()
        mlp = MLPClassifier(hidden_layer_sizes=(50, 40, 40, 80, 20), alpha=a, max_iter=mi)
        mlp.fit(X_train, y_train)
        end = time() - st
        
        acc_tr = accuracy_score(y_train, mlp.predict(X_train)) # Train Accuracy
        acc = accuracy_score(y_test, mlp.predict(X_test)) # Test Accuracy
        df.loc[i] = [a,mi,acc_tr,acc,end]
        i=i+1
```
